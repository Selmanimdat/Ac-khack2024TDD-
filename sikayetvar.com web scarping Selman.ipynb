```python
"""
Eğer sizde kurulu değilse aşağıdaki komut ile çalıştırın
!pip install pandas
pandas-> kütüphane ismi her kütüphane için değiştirin örneğin
!pip install bs4
!pip install selenium
"""

from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.alert import Alert
from undetected_chromedriver import Chrome
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service
import itertools
import pandas as pd
import time
```

```python
# Gecko diriver yolu(gerekirse kullanın)
geckodriver_path = '/usr/local/bin/geckodriver'

# Linkleri toplamak için değişken
linkler = []

# Bütün sayfaları dolaşmak için for döngüsü
for i in range(1, 3):
    # Bütün sayfalar
    url = "https://www.sikayetvar.com/turkcell?page={}".format(i)

    # Geckodriver için servis objesi
    service = Service()

    # Servis ayarları ile driver başlatma
    driver = webdriver.Firefox(service=service)

    try:
        # Ayarlanan url ye git
        driver.get(url)

        # 5 sn sayfanın yüklenemsini bekle
        time.sleep(5)

        # Sayfa kaynağını al
        page_source = driver.page_source

        # bs4 ile parçala 
        soup = BeautifulSoup(page_source, 'html.parser')

        # CSS selector ile linklerin olduğu container ı seç
        links = soup.select('a.complaint-layer.card-v3-container')
        # Containerdaki linkleri al ve linkler listesine ekle
        
        for link in links:
            href = link.get('href')
            print(href)
            linkler.append(href)
        
    finally:
        # En sın çık 
        driver.quit()

# Opsiyonel linkleri yazdır
#print(linkler)
```

```python
# Çekilen linkelrin sayısı
len(linkler)
```

```python
#linkleri metin dosysına kaydetme
my_list = linkler# bizim link listesi
with open("sikayetvar_turkcell_yorum_linkleri.txt", "w") as file:
    for item in my_list:
        file.write("%s\n" % item)
```

```python
# Dosya yolunu belirt
dosya_yolu = 'sikayetvar_turkcell_yorum_linkleri.txt'

# Boş bir liste oluştur
liste = []

# Dosyayı oku ve içeriği listeye ata
with open(dosya_yolu, 'r', encoding='utf-8') as dosya:
    for satir in dosya:
        liste.append(satir.strip())  # strip() boşlukları ve yeni satır karakterlerini kaldırır

```

```python
import random
```

```python
#linklerin çekilcek kadarını alıyoruz
rastgele_elemanlar = random.sample(liste, 500)
textler=[]
```

```python
service = Service()
# Blok reklamları engellemek için firefox ayarı
options = Options()
options.set_preference("dom.disable_open_during_load", True)

# Driver ı başlatıyoruz
driver = webdriver.Firefox(service=service, options=options)

# Ana url
base_url = 'https://www.sikayetvar.com'

for i in rastgele_elemanlar:
    # Her linki dolaşıyoruz
    try:
        # Çektiğimiz yorum linkleri ile ana linkelri birleştiriyoruz
        url = base_url + str(i)
        driver.get(url)
        # Firefox işe yaramazsa diye selenium ile pop upları kapatıyoruz
        try:
            alert = Alert(driver)
            alert.dismiss()  # or alert.accept()
        except:
            pass
        # Sayfa yüklenene kadar bekle
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'complaint-detail-description'))
        )
        # Belirli bir sınıfa sahip elementi bulma
        element = driver.find_element(By.CLASS_NAME, 'complaint-detail-description')

        # Elementin altındaki tüm metinleri alma
        text_content = element.text
        textler.append(text_content)
        print(text_content)
        print("-" * 80)
    except Exception as e:
       # Hata olursa bekleme hatayı yaz geç
        print("The error is: ",e)
        pass
# WebDriver'ı kapatma
driver.quit()
```
