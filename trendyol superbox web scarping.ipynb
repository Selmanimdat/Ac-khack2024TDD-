```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoAlertPresentException
import time

# Links to be scraped
links = ["https://www.trendyol.com/turkcell/4-5g-winn-wifi-superbox-go-mw70vk-siyah-resmi-dist-garantili-p-286249180/yorumlar?boutiqueId=61&merchantId=942102",
        "https://www.trendyol.com/huawei/superbox-b535-232-300-mbps-4-5g-modem-p-104007514/yorumlar?boutiqueId=61&merchantId=164784",
         "https://www.trendyol.com/zte/superbox-plug-play-modem-mf286r-uyumlu-p-282933247/yorumlar?boutiqueId=61&merchantId=260698",
         
]

textler = []

# Geckodriver için servis objesi
service = Service()

# WebDriver'ı başlatma
driver = webdriver.Firefox(service=service)

for url in links:
    driver.get(url)

    try:
        # Handle alert if present
        try:
            alert = driver.switch_to.alert
            alert.dismiss()  # ya da alert.accept()
        except NoAlertPresentException:
            pass

        # Sayfa yüklenene kadar bekle
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'comment-text'))
        )

        SCROLL_PAUSE_TIME = 1

        # Scroll and collect reviews
        while True:
            # Belirli bir sınıfa sahip elementi bulma
            elements = driver.find_elements(By.CLASS_NAME, 'comment-text')

            # Elementin altındaki tüm metinleri alma
            for element in elements:
                text_content = element.text
                if text_content not in textler:
                    textler.append(text_content)
                    print(text_content)
                    print("-" * 80)

            # Scroll down to the bottom of the page
            last_height = driver.execute_script("return document.body.scrollHeight")
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait to load page
            time.sleep(SCROLL_PAUSE_TIME)

            # Calculate new scroll height and compare with last scroll height
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break

    except TimeoutException as e:
        print(f"Timeout occurred while loading {url}: {e}")
    except Exception as e:
        print(f"An error occurred while processing {url}: {e}")

# Close the browser
driver.quit()

# Çekilen metinleri yazdırma
for text in textler:
    print(text)
    print("-" * 80)
```
